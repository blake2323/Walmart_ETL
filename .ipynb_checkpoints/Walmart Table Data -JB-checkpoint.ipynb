```python
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
from pandas.plotting import table
```

```python
features_file = "Resources/features.csv"
train_file = "Resources/train.csv"
```

### No markdown because there's only one year of markdown data...

```python
aa = pd.read_csv(features_file)
bb = aa[['Store', 
                 'Date',
                'Temperature',
                'Fuel_Price',
                'CPI',
                'Unemployment',
                'IsHoliday']]
inital_import_features = bb.head()
inital_import_features #First time importing features.csv into Pandas
```

```python
cc = bb.rename(columns={"Store": "store",
                                "Date": "date",
                                "Temperature": "temperature",
                                "Fuel_Price": "fuel_price",
                                "CPI": "cpi",
                                "Unemployment": "unemployment",
                                "IsHoliday":"is_holiday"})
dd = cc.reset_index()
features_df = dd.rename(columns={"index": "id"})
features_df.head() #Renamed field titles and added id for Primary Key
```

```python
ee = pd.read_csv(train_file)

ff = ee.rename(columns={"Store": "store",
                             "Dept": "dept",
                             "Date": "date",
                             "Weekly_Sales": "weekly_sales",
                             "IsHoliday": "is_holiday"})
gg = ff.reset_index()
train_df = gg.rename(columns={"index": "id"})
train_df.head() #Imported train.csv to Pandas, renamed field titles and added id for Primary Key
```
